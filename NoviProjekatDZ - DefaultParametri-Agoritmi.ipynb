```python
#Odvajamo izlaznu varijablu IsBadBuy i smestamo je u varijablu Y, ostale atribute smestamo u varijablu X
X = data.drop("IsBadBuy", axis = 1)
Y = data['IsBadBuy']
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score, roc_auc_score


#pravimo funkciju evaluateMetrics koja ce racunati accuracy, precision, recall i f1 statistiku u zavisnosti od izlazne varijable y 
#koju vec poznajemo i y_pred koje cemo dobiti predvidjanjem iz nekog modela
def evaluate_metrics(Y, Y_pred):
    print(f'Accuracy : {accuracy_score(Y, Y_pred)}')
    print(f'Precision : {precision_score(Y, Y_pred)}')
    print(f'Recall : {recall_score(Y, Y_pred)}')
    print(f'F1 : {f1_score(Y, Y_pred)}')
```

```python
#delimo dataset na train i test
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=0)
```

```python
print(X_train.describe(include= "all"))
print(type(X_train))
```

```python
print(Y_train.describe(include= "all"))
print(type(Y_train))
```

# Kreiranje modela
Kreiracemo 4 modela:
1. Stablo odlučivanja- klasifikacija
2. KNN- klasifikacija
3. Voting jer stacking nije mogao da ucita paket
4. Bagging

Za način validacije rešenja, koristićemo crossvalidation

```python
X_train = X_train.replace((np.inf, -np.inf, np.nan), 0).reset_index(drop=True)
X_test = X_test.replace((np.inf, -np.inf, np.nan), 0).reset_index(drop=True)
#posto nam je izbacivao ovu gresku Input contains NaN, infinity or a value too large for dtype('float64').
#nakon sredjivanja podataka, pronadjeno je resenje da se inf i nan vrednost zamene 0, svesno unosimo gresku u train set podataka
```

```python
X_train.head(10)
```

```python
#Kreiramo model stabla odlučivanja, koristeći default parametre
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score

tree = DecisionTreeClassifier()
fitted_tree=tree.fit(X_train, Y_train)
```

```python
import sklearn.tree as tr
from matplotlib import pyplot as plt
```

```python
fig = plt.figure(figsize=(25,20))
tree_plotted = tr.plot_tree(fitted_tree, max_depth=3,
                   filled=True)
#Pitati kako da uzmemo feature_names iz dataseta
```

```python
#koristimo 10tostruku kors validaciju za proveru koliko je model dobar
# kada prodjemo kobinaciju 10 foldova nad x i y train podacimam izracunacemo srednju gresku koja je napravljena
#kros validacijom dobijamo prosecnu vrednost dobijanog auc parametra kroz deset iteracija
cross_val_score(tree ,X_train, Y_train, cv = 10, scoring="roc_auc").mean()
```

```python
#pozivamo predikcije nad test podacima, koji predvidja koja observacija pripada kojoj klasi, 0 ili 1 
Y_pred_tree = tree.predict(X_test)
Y_pred_tree
```

```python
confusion_matrix(Y_test, Y_pred_tree)
#TN [0,0]
#FN [1,0]
#TP [1,1]
#FP[0,1]

#true/ predicted
#NO     NO      YES
       #TN      FP
#YES   FN      TP

# 1503 automobila su po modelu dobro svrstana u dobru kupovinu, dok je 46 svsrtano u losu kupovinu
# 206 automobila je svrstano u dobru kupovinu, a zapravo ce biti losa kupovina
#dok je 285 automobila svrstano u losu kupovinu a zapravo bi bili dobra kupovina
```

```python
#za dobijenu matricu konfuzije racunamo evaluacione metrike
evaluate_metrics(Y_test, Y_pred_tree)
```

```python
#Mozemo zakljuciti da su metrike poprilicno lose i da treba popraviti indentifikaciju pozitivne klase tj necemo da gresimo kod FN
#Necemo da nas laze da je nesto FN nego da nam stvarno svrsta u TP, znaci treba da poboljsamo recall
#Za stablo odlucivanja cemo optimizovati parametre i uporediti metrike novog modela i ovog modela sa default parametrima
```

```python
#Kreiramo model KNN
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn.fit(X_train, Y_train)
```

```python
cross_val_score(knn, X_train, Y_train, cv = 10, scoring="roc_auc").mean()
```

```python
Y_pred_knn = knn.predict(X_test)
```

```python
confusion_matrix(Y_test, Y_pred_knn)
```

```python
#Racunamo evaluacione metrike za knn matricu konfuzije
evaluate_metrics(Y_test, Y_pred_knn)
#Vidimo da nam je precision bolji nego kod stabla odlucivanja jer je pronasao manje FP vrednosti, 26, dok je kod drva bilo 285
# Recall nam je losiji nego kod drva odlucivanja, jer smo imali dosta FN, sto znaci da nas je dosta slagao da kupovina nije losa
#Tj rekao nam je da je kupovina dobra, medjutim kupovina je losa, sto nije dobro za poslovanje
```

```python
#Kreiramo Voting model, a kao ulazne parametre mu prosledjujemo kreirane modele drvo i knn na osnovu kojih on donosi svoje zakljucke
from sklearn.ensemble import VotingClassifier
```

```python
voting_model = VotingClassifier(voting = "soft", estimators=[('tree', tree), ('knn', knn)])

voting_model.fit(X_train, Y_train)
```

```python
cross_val_score(voting_model, X_train, Y_train, cv=10, scoring='roc_auc').mean()
```

```python
Y_pred_voting = voting_model.predict(X_test)
confusion_matrix(Y_test, Y_pred_voting)
```

```python
evaluate_metrics(Y_test, Y_pred_voting)
```

```python
#Kreiramo Bagging model
from sklearn.ensemble import BaggingClassifier
model_bagging = BaggingClassifier(base_estimator = knn, bootstrap=100)
model_bagging.fit(X_train, Y_train)
```

```python
cross_val_score(model_bagging, X_train, Y_train, cv=10, scoring='roc_auc').mean()
```

```python
Y_pred_bagging = model_bagging.predict(X_test)
confusion_matrix(Y_test, Y_pred_bagging)
```

```python
evaluate_metrics(Y_test, Y_pred_bagging)
```

```python
#Kada sumiramo evaluacione metrike za sva cetiri modela, ukoliko posmatramo accuracy pokazatelj najbolji model je Bagging
```
