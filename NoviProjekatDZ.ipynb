# Standardizacija i dummy coding

```python
#S obzirom da object varijable treba da pretvorimo u nesto smisleno, tj faktorske/kategoricke, primenjujemo dummz coding
#Uvidjamo da varijable Model sa 632, Trim sa 104, Submodel sa 465 unique vrednosti nisu pogodne za dummy coding koje bismo koristili kod linearne regresije
#Takodje ovim varijablama ne mozemo napraviti poredak, tako da ne mogu postati faktorske, stoga ih izbacujemo
#Numericke varijable cemo standardizovati, jer imaju autlajere pa ne mozemo primeniti normalizaciju
#A ostale object varijable cemo srediti preko dummy codinga jer njihove klase nisu medjusobno uporedive
```

```python
data1 = data1.drop(['Model', 'Trim', 'SubModel'], axis = 1)
```

```python
print(data1.describe(include= "all"))
```

```python
# u X cemo smestiti sve ulazne varijable (prediktore)
# u Y cemo smestiti izlaznu var (IsBadBuy)
Y = data1['IsBadBuy']
X = data1.drop("IsBadBuy", axis = 1)
```

```python
#Sredimo nase podatke pomocu ColumnTransforema-a. 
#Pomocu StandarScaler-a cemo standardizovati sve numericke varijable 
#a pomocu OneHotEncoder-a cemo nase kategorijske varijable pretvoriti u dummy varijable, 
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer 
from sklearn.preprocessing import StandardScaler
```

```python
#Izdvojicemo indekse numerickih i kategorickih koloni u data setu
#Onda cemo posebno na numericke posebno na kategoricke kolone primenite funnkcije za standardizaciju i dummy coding
#Zatim cemo napraviti novi data frame sa svim podacima koji ce sadrzati sve standardizovane numericke kolone
#kao i sve nove kolone kategorickih varijabli koje smo dobili prilikom transformisanja klasa
#Svaka klasa jedne kolone je postala nova kolona za sebe za vrednosti 0 ili 1 u zavisnosti od pripadnosti
```

```python
numerics = X.select_dtypes(np.number).columns
numeric_transformer = StandardScaler()

categorical = X.select_dtypes(object).columns
categorical_transformer = OneHotEncoder(sparse=False)

preprocessor = ColumnTransformer(
    transformers=[
        ('numbers', numeric_transformer, numerics),
        ('categorical', categorical_transformer, categorical)
    ]
)

preprocessor.fit(X)
```

```python
X = pd.DataFrame(preprocessor.transform(X), 
             columns=[*numerics, *preprocessor.transformers_[1][1].get_feature_names(categorical)])
```
